```python
findal_df = pd.concat([df,pd.get_dummies(df['Department'],drop_first=True)],axis=1)
```

```python
findal_df
```

```python
findal_df.drop('Department',axis=1,inplace=True)
```

```python
findal_df
```

```python
final_df = findal_df
```

```python
X  = final_df.drop('left',axis=1)
```

```python
y = final_df['left']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
le = LogisticRegression(max_iter=500)
```

```python
model = le.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test,y_pred)
```
